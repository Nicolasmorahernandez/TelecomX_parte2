```python
# === 1) Matriz de correlación (solo numéricas) ===
num = df_tratado.drop(columns=['Churn'], errors='ignore').select_dtypes(include=[np.number])
corr = num.corr()  # Pearson

# === 2) Listar pares con |correlación| > umbral ===
umbral = 0.8  # ajusta a gusto
pairs = (
    corr.where(np.triu(np.ones(corr.shape), k=1).astype(bool))   # triángulo superior
        .stack()
        .rename('corr')
        .abs()
        .sort_values(ascending=False)
)
altas = pairs[pairs > umbral]
print("Pares con |r| >", umbral)
print(altas.head(50))  # quita el head si quieres todo

# === 3) Elegir automáticamente columnas a eliminar por redundancia ===
# Estrategia: para cada par, eliminar la columna con mayor "correlación media absoluta" con el resto.
if not altas.empty:
    mean_abs_corr = corr.abs().mean()
    drop_cols = set()
    for (c1, c2), r in altas.items():
        # si ya marcamos una de las dos, salta
        if c1 in drop_cols or c2 in drop_cols:
            continue
        cand = c1 if mean_abs_corr[c1] >= mean_abs_corr[c2] else c2
        drop_cols.add(cand)

    print("\nSugerencia de columnas a eliminar por multicolinealidad:")
    print(sorted(drop_cols))

    # Si quieres aplicar la eliminación:
    # df_reducido = df_tratado.drop(columns=list(drop_cols))
else:
    print("\nNo se detectaron pares por encima del umbral.")
```

Comportamiento: Más conservador, prefiere “estar seguro” antes de marcar un cliente como churn, pero deja escapar casos reales.

2. Identificación de Overfitting o Underfitting

Logistic Regression:
El hecho de que tenga un recall alto pero precisión moderada indica que el modelo puede estar ajustado para sensibilidad. Esto no es overfitting en sí, pero podría indicar que está sobreponderando la clase minoritaria debido al balanceo, aumentando falsos positivos.

Random Forest:
El bajo recall y una precisión moderada sugieren que podría estar subajustado (underfitting), posiblemente por una configuración conservadora (número de árboles, profundidad máxima) o porque el balanceo no fue óptimo para este tipo de modelo.

3. Posibles mejoras

Afinar hiperparámetros:

Para Logistic Regression: Ajustar C y probar diferentes métodos de regularización (l1, l2, elasticnet).

Para Random Forest: Incrementar profundidad máxima (max_depth), número de árboles (n_estimators), y ajustar class_weight más agresivamente.

Feature Engineering:

Explorar interacciones entre variables.

Probar discretización de variables continuas con alto impacto.

Umbral de decisión:

Ajustar el threshold de predicción para mejorar la relación entre precisión y recall según el objetivo de negocio (minimizar falsos negativos o falsos positivos).

4. Recomendación según objetivo

Si el objetivo es detectar la mayor cantidad posible de clientes en riesgo (priorizando recall), usaría Logistic Regression con threshold ajustado.

Si el objetivo es evitar gastar recursos en falsos positivos, Random Forest puede ser preferible, pero habría que mejorar su recall.

```python
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score, precision_recall_fscore_support, roc_auc_score

def eval_model(nombre, modelo, X, y):
    """Evalúa un modelo y devuelve métricas en un dict."""
    y_pred = modelo.predict(X)
    # Probabilidades / score para AUC
    y_score = None
    if hasattr(modelo, "predict_proba"):
        y_score = modelo.predict_proba(X)[:, 1]
    elif hasattr(modelo, "decision_function"):
        y_score = modelo.decision_function(X)

    acc = accuracy_score(y, y_pred)
    prec, rec, f1, _ = precision_recall_fscore_support(
        y, y_pred, average='binary', pos_label=1, zero_division=0
    )
    auc = roc_auc_score(y, y_score) if y_score is not None else np.nan

    return {
        "Modelo": nombre,
        "Accuracy": acc,
        "Recall (Churn=1)": rec,
        "Precision (Churn=1)": prec,
        "F1 (Churn=1)": f1,
        "AUC": auc
    }

# ------- Selección de modelos y el X_test correcto -------
registros = []

# 1) Regresión Logística (usa datos normalizados)
registros.append(eval_model("LogReg (scaled, balanced)", modelo_log, X_test_norm, y_test))

# 2) Random Forest (sin normalizar)
registros.append(eval_model("RandomForest (balanced)", modelo_rf, X_test, y_test))

# 3) Decision Tree (sin normalizar)
registros.append(eval_model("DecisionTree (balanced)", modelo_tree, X_test, y_test))

# 4) KNN (usa datos normalizados) — toma best_knn si existe; si no, knn_bal
try:
    registros.append(eval_model("KNN (scaled)", best_knn, X_test_norm, y_test))
except NameError:
    registros.append(eval_model("KNN (scaled)", knn_bal, X_test_norm, y_test))

# ------- Tabla final ordenada por AUC (o por Recall si prefieres) -------
tabla = pd.DataFrame(registros)
tabla = tabla.round(3).sort_values(["AUC", "Recall (Churn=1)"], ascending=False).reset_index(drop=True)
print(tabla)
```
